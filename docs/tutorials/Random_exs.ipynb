```python
%autoreload
plt.style.use("ggplot")
scales = ["PAQ1", "PAQ2", "PAQ3", "PAQ4", "PAQ5", "PAQ6", "PAQ7", "PAQ8"]

ssm_res = circumplex.ssm_analyze(
    data, scales, angles=octants(), grouping="Language", measures=["loud"]
)
fig = ssm_res.plot_circle()
plt.show()
```
